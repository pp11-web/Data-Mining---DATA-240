# Question 1 EDA Graph Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Dataset I': {
        'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
        'y': [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
    },
    'Dataset II': {
        'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
        'y': [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
    },
    'Dataset III': {
        'x': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
        'y': [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
    }
}
```

```python
df1 = pd.DataFrame(data['Dataset I'])
df2 = pd.DataFrame(data['Dataset II'])
df3 = pd.DataFrame(data['Dataset III'])
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(df1['x'], df1['y'], color='blue')
plt.title('Dataset I')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 3, 2)
plt.scatter(df2['x'], df2['y'], color='green')
plt.title('Dataset II')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 3, 3)
plt.scatter(df3['x'], df3['y'], color='red')
plt.title('Dataset III')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(3, 3, 1)
plt.hist(df1['y'], color='blue', bins=10, edgecolor='black')
plt.title('Dataset I - Histogram of y')
plt.xlabel('y')
plt.ylabel('Frequency')

plt.subplot(3, 3, 2)
plt.hist(df2['y'], color='green', bins=10, edgecolor='black')
plt.title('Dataset II - Histogram of y')
plt.xlabel('y')
plt.ylabel('Frequency')

plt.subplot(3, 3, 3)
plt.hist(df3['y'], color='red', bins=10, edgecolor='black')
plt.title('Dataset III - Histogram of y')
plt.xlabel('y')
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(3, 3, 4)
plt.boxplot(df1['y'], patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title('Dataset I - Boxplot of y')

plt.subplot(3, 3, 5)
plt.boxplot(df2['y'], patch_artist=True, boxprops=dict(facecolor='green'))
plt.title('Dataset II - Boxplot of y')

plt.subplot(3, 3, 6)
plt.boxplot(df3['y'], patch_artist=True, boxprops=dict(facecolor='red'))
plt.title('Dataset III - Boxplot of y')
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(3, 3, 7)
plt.plot(df1['x'], df1['y'], marker='o', color='blue')
plt.title('Dataset I - Line Plot')
plt.xlabel('x')
plt.ylabel('y')

# Line plot for Dataset II
plt.subplot(3, 3, 8)
plt.plot(df2['x'], df2['y'], marker='o', color='green')
plt.title('Dataset II - Line Plot')
plt.xlabel('x')
plt.ylabel('y')

# Line plot for Dataset III
plt.subplot(3, 3, 9)
plt.plot(df3['x'], df3['y'], marker='o', color='red')
plt.title('Dataset III - Line Plot')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```

Explainnation

Dataset II shows a concentration of values near 9, with a skew toward lower values.

Dataset III shows a concentration of values between 6 and 8, with one outlier around 12.

Boxplots:

Dataset I has a moderate spread, with no outliers.

Dataset II shows a higher y-range, with one outlier below 4.

Dataset III has a narrower spread, with one outlier above 12.

Line Plots:

Dataset I shows a non-linear, somewhat jagged relationship between x and y.

Dataset II has a curvilinear shape, indicating a peak in the middle.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
file_path = 'data_for_pca.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
# Task (a): Compute the correlation matrix for the data excluding the output variable 'y'
correlation_matrix = data.drop(columns=['y']).corr()
correlation_matrix
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X = data.drop(columns=['y'])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Task (b): Perform PCA
pca = PCA(n_components=len(X.columns))
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(8, 6))
plt.plot(np.arange(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='x', label='Cumulative Variance')
plt.title('Scree Plot: Explained Variance by Principal Components')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Identify the components with eigenvalues >= 1.0
eigenvalues = pca.explained_variance_
components_to_keep = [i for i, val in enumerate(eigenvalues) if val >= 1.0]
```

```python
X_reduced = X_pca[:, components_to_keep]
```

```python
reduced_data = pd.DataFrame(X_reduced)
reduced_data.head(10)
```
